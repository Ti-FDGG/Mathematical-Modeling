### 基本设置

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

plt.rcParams['font.sans-serif'] = ['SimHei']  #显示中文
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
excel = pd.read_excel('第二题数据.xlsx')
lon_lat = pd.read_excel('等高线图.xlsx')
data = excel.iloc[0:,2:]    
ana_ff1 = pd.DataFrame(data)
```

### 全部指标地区分布折线图

```python
x_label = excel['断面名称']
y_labels = list(excel.columns[2:])
# print(y_labels)

plt.figure(figsize=[27.50, 12.50])  

# 遍历y_labels字典，为每个指标创建子图  
for i, ylabel in enumerate(y_labels, start=1):  
    plt.subplot(3, 3, i)  # 根据循环的索引选择子图位置  
    plt.plot(x_label, excel[ylabel], label=ylabel, marker='o')  
    plt.xticks(rotation=40, fontsize=12) 
    plt.title(ylabel,fontsize=20)  
    plt.gca().xaxis.set_tick_params(labelrotation=40)

# plt.legend()  
plt.tight_layout() 

plt.show()
```
