```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
df1 = pd.read_excel('../excels/2.xlsx', sheet_name='2023年的农作物种植情况')
df1.fillna(method='ffill', inplace=True)
# print(df1)
df2 = pd.read_excel('../excels/2.xlsx', sheet_name='2023年统计的相关数据')
# print(df2)
df2023 = pd.read_excel('../excels/附件3/2023.xlsx', sheet_name='2023')
# 将读取到的表格中的NaN取值为0
df2023.fillna(0, inplace=True)
print(df2023)
```

```python
# 将 '单季' 替换为 '第一季'
df1['种植季次'] = df1['种植季次'].replace('单季', '第一季')

# 遍历 df1 的每一行
for index, row in df1.iterrows():
    season = row['种植季次']
    field = row['种植地块']
    crop_name = row['作物名称']
    area = row['种植面积/亩']
    
    # 找到 df2023 中对应的行
    df2023_row = df2023[(df2023['种植季次'] == season) & (df2023['种植地块'] == field)]
    
    if not df2023_row.empty:
        # 找到对应的列
        if crop_name in df2023.columns:
            df2023.loc[df2023_row.index, crop_name] = area

# 打印 df2023
print(df2023)

# 将 df2023 保存到2023.xlsx的Sheet1中
df2023.to_excel('../excels/附件3/2023_result.xlsx', index=False)
```
